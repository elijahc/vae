```python
class GAN():
    def __init__(self,G,D,C):
        self.G = G
        self.D = D
        self.C = C
        self.AM = None
        self.GM = None
        
    def build_AM(self):
        self.C.trainable=False
        self.D.trainable=False
        G_input = Input(shape=self.G.get_input_shape_at(1)[1:],name='AM_input')
        G_out = self.G(G_input)
        self.AM = Model(
            inputs=G_input,
            outputs=[self.D(G_out),self.C(G_out)],
            name='adversarial'
        )
        losses={
            'D': 'binary_crossentropy',
            'C': 'categorical_crossentropy'
        }
        metrics={
            'D': 'accuracy',
            'C': 'accuracy'
        }
        self.AM.compile(optimizer='adam',loss=losses,metrics=metrics)
        return self.AM
    
    def build_DM(self):
        self.C.trainable=True
        self.D.trainable=True
        self.G.trainable=False
        DM_input = Input(shape=self.C.get_input_shape_at(0)[1:],name='DM_input')
        c_out = self.C(DM_input)
        d_out = self.D(DM_input)
        z_out = self.
        self.DM = Model(
            inputs=DM_input,
            outputs=[self.D(DM_input),self.C(DM_input)],
            name='discriminator'
        )
        losses={
            'D': 'binary_crossentropy',
            'C': 'categorical_crossentropy'
        }
        metrics={
            'D': 'accuracy',
            'C': 'accuracy'
        }
        self.DM.compile(optimizer='adam',loss=losses,metrics=metrics)
        return self.DM
```
