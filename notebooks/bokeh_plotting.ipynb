```python
import numpy as np

from bokeh.io import output_notebook,show
from bokeh.layouts import gridplot
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource

from src.results.dicarlo import get_dicarlo_su
```

```python
output_notebook()
```

```python
proj_root = '/home/elijahc/projects/vae'
```

```python
dicarlo_df = get_dicarlo_su(proj_root,fn='su_selectivity_dicarlo_hi_var.pqt')
dicarlo_df = dicarlo_df.fillna(0).query('region != "pixel"')
```

```python
source = ColumnDataSource(dicarlo_df.query('region=="IT"')[['category_name','tx','ty']])
```

```python
TOOLS = "box_select,lasso_select,reset,help"

ct,edges = np.histogram(dicarlo_df.category_name,bins=20)
s1 = figure(tools=TOOLS,plot_width=500, plot_height=500, title=None)
s1.cross(x='tx',y='ty',size=8,source=source)

ct,edges = np.histogram(dicarlo_df.tx,bins=20)
s2 = figure(tools=TOOLS,plot_width=500, plot_height=500, title=None)
s2.quad(left=edges[:-1],right=edges[1:],top=ct,bottom=0,)
```

```python
p = gridplot([[s1,s2]])
show(p)
```

```python
x = list(range(11))
y0 = x
y1 = [10-xx for xx in x]
y2 = [abs(xx-5) for xx in x]

# create a new plot
s1 = figure(plot_width=250, plot_height=250, title=None)
s1.circle(x, y0, size=10, color="navy", alpha=0.5)

# create a new plot and share both ranges
s2 = figure(plot_width=250, plot_height=250, x_range=s1.x_range, y_range=s1.y_range, title=None)
s2.triangle(x, y1, size=10, color="firebrick", alpha=0.5)

# create a new plot and share only one range
s3 = figure(plot_width=250, plot_height=250, x_range=s1.x_range, title=None)
s3.square(x, y2, size=10, color="olive", alpha=0.5)

p = gridplot([[s1, s2, s3]], toolbar_location=None)

# show the results
show(p)
```
